```python
import pandas as pd
```

```python
schedule = pd.read_csv('2015schedule.csv')
```

```python
schedule.head()
```

```python
t1 = schedule['Visitor/Neutral'].values
t2 = schedule['Home/Neutral'].values
teams = list(set(t1) | set(t2))
```

```python
team_numgames = {}
for i in range(len(t1)):
    name1arr = t1[i].split()
    name2arr = t2[i].split()
    name1 = name1arr[len(name1arr)-1]
    name2 = name2arr[len(name2arr)-1]
    tupe = (name1, name2) if name1 < name2 else (name2, name1)
    if tupe in team_numgames:
        team_numgames[tupe] += 1
    else:
        team_numgames[tupe] = 1
team_numgames
```

```python
team_confopponents = dict(zip(teams, [set([])]*30))
```

```python
for i in range(len(t1)):
    team_confopponents[t1[i]].add(t2[i])
    team_confopponents[t2[i]].add(t1[i])
```

```python
team_confopponents.keys()[0].split()
```

```python
team_abbrev = {}
for k in team_confopponents:
    arr = k.split()
    tname = arr[len(arr)-1]
    team_abbrev[tname] = set([])
    for opp in team_confopponents[k]:
        arr = opp.split()
        oppname = arr[len(arr)-1]
        team_abbrev[tname].add(oppname)
team_abbrev
```

```python
d = datagen.DataGen(2015)
d
```

```python
t_conf = {}
for t in team_abbrev:
    team = d.league.get_team("Trailblazers" if t == "Blazers" else t)
    t_conf["Trailblazers" if t == "Blazers" else t] = set([])
    for opp in team_abbrev[t]:
        tupe = (t,opp) if t < opp else (opp, t)
        oppteam = d.league.get_team("Trailblazers" if opp == "Blazers" else opp)
        if oppteam.division is not team.division and oppteam.division.conference is team.division.conference and team_numgames[tupe] == 4:
            t_conf["Trailblazers" if t == "Blazers" else t].add("Trailblazers" if opp == "Blazers" else opp)
t_conf
for k in t_conf:
    if len(t_conf[k]) != 6:
        print k
```
